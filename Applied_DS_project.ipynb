##### Older people are not as tech-savvy and probably prefer shopping in-store. 

```python
import pandas as pd
from scipy import stats

# Load your CSV data into a DataFrame (assuming your CSV file is named 'data.csv')


# Split the data into two groups: people aged 50 or older and people younger than 50
older_age = df[df['AGE'] >= 50]
web_shopping = older_age['NUMWEBPURCHASES']
store_shopping = older_age['NUMSTOREPURCHASES']
# Perform the t-test
_, p_value = stats.ttest_ind(store_shopping,web_shopping,alternative="greater")
# Set the significance level (alpha)
alpha = 0.05

# Print the results
#print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Compare the p-value to alpha and make a conclusion
if p_value <= alpha:
    print("Reject the null hypothesis: lder people are not as tech-savvy and probably prefer shopping in-store.")
else:
    print("Fail to reject the null hypothesis: older peole are happy with shopping in store.")
```

##### Customers with kids probably have less time to visit a store and would prefer to shop online.

```python
#NUMBER_CHILDREN 

import pandas as pd
from scipy import stats

# Load your CSV data into a DataFrame (assuming your CSV file is named 'data.csv')


# Split the data into two groups: people aged 50 or older and people younger than 50
kids = df[df['NUMBER_CHILDREN'] > 0]
web_shopping = kids['NUMWEBPURCHASES']
store_shopping = kids['NUMSTOREPURCHASES']
# Perform the t-test
_, p_value = stats.ttest_ind(store_shopping,web_shopping,alternative="greater")
# Set the significance level (alpha)
alpha = 0.05

# Print the results
#print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Compare the p-value to alpha and make a conclusion
if p_value <= alpha:
    print("Reject the null hypothesis: People with children are more likely to do shopping through web stores.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in shopping behavior between people with and without children.")
```

##### Other distribution channels may cannibalize sales at the store.

```python
import pandas as pd
from scipy import stats

# Load your data into the 'df' DataFrame

# Define your groups
web_shoppers = df[df['NUMWEBPURCHASES'] > 0]
catalog_shoppers = df[df['NUMCATALOGPURCHASES'] > 0]
in_store_shoppers = df[df['NUMSTOREPURCHASES'] > 0]

# Perform t-tests
_, p_value_web = stats.ttest_ind(in_store_shoppers['NUMSTOREPURCHASES'], web_shoppers['NUMSTOREPURCHASES'])
_, p_value_catalog = stats.ttest_ind(in_store_shoppers['NUMSTOREPURCHASES'], catalog_shoppers['NUMSTOREPURCHASES'])

# Set the significance level (alpha)
alpha = 0.05

# Print the results and make conclusions
print("Web vs. In-Store:")
print("p-value:", p_value_web)
if p_value_web <= alpha:
    print("Reject the null hypothesis: Web purchases may cannibalize in-store sales.")
else:
    print("Fail to reject the null hypothesis: No significant difference observed.")

print("\nCatalog vs. In-Store:")
print("p-value:", p_value_catalog)
if p_value_catalog <= alpha:
    print("Reject the null hypothesis: Catalog purchases may cannibalize in-store sales.")
else:
    print("Fail to reject the null hypothesis: No significant difference observed.")
```

#### Does the US fare significantly better than the rest of the world in terms of total purchases?

```python
import pandas as pd
from scipy import stats

# Load your data into the 'df' DataFrame

# Separate the data into U.S. customers and rest of the world customers
us_customers = df[df['COUNTRY'] == 'us']
rest_of_world_customers = df[df['COUNTRY'] != 'us']

# Perform a t-test to compare total purchases between U.S. and rest of the world customers
_, p_value = stats.ttest_ind(us_customers['TOTAL_SPENDING'], rest_of_world_customers['TOTAL_SPENDING'])

# Set the significance level (alpha)
alpha = 0.05

# Print the results and make a conclusion
print("p-value:", p_value)
if p_value <= alpha:
    print("Reject the null hypothesis: The United States fares significantly better in terms of total purchases.")
else:
    print("Fail to reject the null hypothesis: No significant difference observed.")
```
